```python
from deepface import DeepFace
import cv2
import matplotlib.pyplot as plt
```

```python
# img_path = "dataset/img11.jpg"
img_path = "../../test.jpeg"
```

```python
img = cv2.imread(img_path)
```

```python
face_objs = DeepFace.extract_faces(img_path=img_path, detector_backend="retinaface")
```

```python
facial_area = face_objs[0]["facial_area"]

x = facial_area["x"]
y = facial_area["y"]
w = facial_area["w"]
h = facial_area["h"]

le = facial_area["left_eye"]
re = facial_area["right_eye"]
```

```python
img = cv2.rectangle(img, (x, y), (x + w, y + h), (67, 67, 67), 1)

img = cv2.circle(img, (le), radius=5, color=(0, 0, 255), thickness=-1) # red
img = cv2.circle(img, (re), radius=5, color=(255, 0, 0), thickness=-1) # blue
```

```python
fig = plt.figure(figsize=(10, 10))

plt.imshow(img[:,:,::-1])
plt.show()
```
