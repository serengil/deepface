```python
from deepface import DeepFace
import matplotlib.pyplot as plt
import cv2
```

```python
img_path = "dataset/img11.jpg"
```

```python
results = DeepFace.represent(img_path, model_name="Facenet", return_face=True)
```

```python
results[0]["face"].shape
```

```python
results = DeepFace.represent(
    ["dataset/img1.jpg", "dataset/img2.jpg", "dataset/couple.jpg"],
    model_name="Facenet",
    return_face=True
)
```

```python
for inner_result in results:
    for result in inner_result:
        face = result["face"]
        print(face.shape)
```
