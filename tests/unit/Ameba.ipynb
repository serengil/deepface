```python
from typing import Optional
```

```python
messages = [
    {
        "reply_to_id": 123,
        "content": "dummy",
        "id_of_message": 456,
    },
    {
        "reply_to_id": None,
        "content": "dummy",
        "id_of_message": 123,
    }
]
```

```python
def find_score(id_of_message: int):
    return 80
```

```python
scores = []
# messages -> 2M

for message in messages:
    score = find_score(message["id_of_message"])
    scores.append(score)

# O(n)
max(scores)
# O(n x logn)
```

```python
pivot = {}
for message in messages:
    pivot[message["id_of_message"]] = message
```

```python
pivot[456]
```

```python
def find_message(message_id: int, frame_size: int = 3):
    chain = []

    current_frame = 0
    current_message_id = message_id
    while True:
        current_frame += 1

        message = pivot.get(current_message_id)

        if message is None:
            raise ValueError(f"{current_message_id} not found")

        chain.append(message)

        # check its parent message
        if message.get("reply_to_id") is not None:
            # parent_message = pivot[message.get("reply_to_id")]
            # chain.append(parent_message)
            current_message_id = message.get("reply_to_id")
        else:
            break

        if current_frame == frame_size:
            break

    return chain
```
